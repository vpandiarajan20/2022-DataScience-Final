```python
# plotting number one songs over time
billboard_data = pd.read_csv("data/billboardHot100_1999-2019.csv")
billboard_data.drop(columns = "Lyrics", inplace=True)
```

```python
billboard_data.head(200)
```

```python
# loading and cleaning billboard
billboardHot100 = pd.read_csv('data/billboardHot100_1999-2019.csv')
billboardHot100 = billboardHot100[['Artists', 'Name', 'Genre', 'Peak.position', 'Weeks.on.chart','Weekly.rank','Date']]
billboardHot100['Weeks.on.chart'] = billboardHot100['Weeks.on.chart'].fillna(1)
billboardHot100['Peak.position'] = billboardHot100['Peak.position'].fillna(100)
billboardHot100['Weekly.rank'] = billboardHot100['Weekly.rank'].fillna(100)
billboardHot100['Genre'] = billboardHot100['Genre'].fillna('None')
billboardHot100['Date'] = billboardHot100['Date'].fillna('1999')
billboardHot100 = billboardHot100.groupby(['Artists', 'Name','Genre']).agg({'Peak.position':'min', 'Weeks.on.chart':'max', 'Weekly.rank':'mean', 'Date':'max'})[['Peak.position','Weeks.on.chart','Weekly.rank', 'Date']].reset_index()
billboardHot100.rename(columns = {'Weekly.rank':'Avg.rank', 'Artists':'Artist'}, inplace = True)
billboardHot100['Date'] = pd.to_datetime(billboardHot100['Date'])
billboardHot100['Date'] = billboardHot100['Date'].apply(lambda x : x.year)
billboardHot100 = billboardHot100.applymap(lambda s: s.upper() if type(s) == str else s)
billboardHot100.head()
```

```python
num_one_names = one_songs_grammy_removed.Name
years = []
for name in num_one_names:
    billboard_song = billboardHot100[billboardHot100.Name == name]
    years.append(billboard_song.Date.iloc[0])
print(years)
```

```python
one_songs_grammy_removed["Year"] = years
one_songs_grammy_removed.head()
```

```python
one_danceability = one_songs_grammy_removed.Danceability
one_energy = one_songs_grammy_removed.Energy
one_valence = one_songs_grammy_removed.Valence
one_year = one_songs_grammy_removed.Year
fig, ax = plt.subplots(3,1, figsize=(6,12))
ax[0].scatter(one_year,one_danceability)
ax[1].scatter(one_year,one_energy)
ax[2].scatter(one_year,one_valence)   

print(ss.pearsonr(one_danceability, one_year))
print(ss.pearsonr(one_energy, one_year))
print(ss.pearsonr(one_valence, one_year))
```

```python
# compare differences between songs who are on the ranking for more than 10 weeks, 
# songs that do not last for more than 10
on_chart_long = df_all[df_all["Weeks.on.chart"] >= 10]
on_chart_short = df_all[df_all["Weeks.on.chart"] < 10]

long_danceability = on_chart_long['Danceability']
long_danceability = [x for x in long_danceability if x > 0 ]

short_danceability = on_chart_short['Danceability']
short_danceability = [x for x in short_danceability if x > 0 ]

print(ss.ttest_ind(short_danceability,long_danceability))

# look at the sentiment of songs
# see if song of year sentiment has changed over time 
```

```python
num_weeks = [5, 10, 15, 20, 25, 30]
for weeks in num_weeks:
    on_chart_long = df_all[df_all["Weeks.on.chart"] >= weeks]
    on_chart_short = df_all[df_all["Weeks.on.chart"] < weeks]
    
    print(len(on_chart_long), len(on_chart_short))
    
    long_danceability = on_chart_long['Danceability']
    long_danceability = [x for x in long_danceability if x > 0 ]

    short_danceability = on_chart_short['Danceability']
    short_danceability = [x for x in short_danceability if x > 0 ]

    print(ss.ttest_ind(short_danceability,long_danceability))
```

```python
num_weeks = [5, 10, 15, 20, 25, 30]
for weeks in num_weeks:
    on_chart_long = df_all[df_all["Weeks.on.chart"] >= weeks]
    on_chart_short = df_all[df_all["Weeks.on.chart"] < weeks]
    
    print("lengths (long, short): ", len(on_chart_long), len(on_chart_short))
    
    long_energy = on_chart_long['Energy']
    long_energy = [x for x in long_energy if x > 0 ]

    short_energy = on_chart_short['Energy']
    short_energy = [x for x in short_energy if x > 0 ]
    
    print("means (long, short): ", np.mean(long_energy), np.mean(short_energy))

    print(ss.ttest_ind(short_energy,long_energy))
    print("\n")
```

```python
num_weeks = [5, 10, 15, 20, 25, 30]
for weeks in num_weeks:
    on_chart_long = df_all[df_all["Weeks.on.chart"] >= weeks]
    on_chart_short = df_all[df_all["Weeks.on.chart"] < 5]
    
    print("lengths (long, short): ", len(on_chart_long), len(on_chart_short))
    
    long_valence = on_chart_long['Valence']
    long_valence = [x for x in long_valence if x > 0 ]

    short_valence = on_chart_short['Valence']
    short_valence = [x for x in short_valence if x > 0 ]
    
    print("means (long, short): ", np.mean(long_valence), np.mean(short_valence))

    print(ss.ttest_ind(short_valence,long_valence))
    print("\n")
```
